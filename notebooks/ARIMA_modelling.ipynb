```python
import sys
sys.path.insert(1,'../src')

import numpy as np, pandas as pd
import warnings
import matplotlib.pyplot as plt
from models import get_ARIMA

# plt.rcParams.update({'figure.figsize':(16, 8), 'figure.dpi':300})
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../data/prices250.csv")
pct = pd.read_csv("../data/prices250_percentage.csv")
```

```python
current_df = df.iloc[0:125, :]
models_dict = get_ARIMA(current_df)
```

```python
key = "98"
size = 100
model_fit = models_dict[key]
pred_df = df.iloc[0:125 + size, :]
fc, se, conf = model_fit.forecast(size, alpha=0.05)  # 95% conf
plt.plot(range(len(current_df) - 1, len(current_df) - 1 + size), fc, label='forecast')
plt.plot(pred_df[key])
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
for key in pct.columns:
    plt.plot(df[key], color="b")
    plt.title("Instrument %s" % key)
    plt.show()
```
